# 📊 Neural Policy Visualizer: PPO-CoordConv-Snake

Evaluate a trained **PPO-CoordConv-Snake** agent with deterministic policy and render a rich telemetry dashboard directly to MP4.

🔗 **GitHub:** [siddhartha399/PPO-CoordConv-Snake](https://github.com/siddhartha399/PPO-CoordConv-Snake)


```python
import os, sys, subprocess
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Clone repo if needed
REPO_URL = 'https://github.com/siddhartha399/PPO-CoordConv-Snake.git'
REPO_NAME = 'PPO-CoordConv-Snake'
if not os.path.exists(REPO_NAME):
    print(f'[INFO] Cloning {REPO_URL}...')
    subprocess.run(['git', 'clone', REPO_URL], check=True)
else:
    print('[OK] Repository already cloned.')

os.chdir(REPO_NAME)
if os.path.exists('requirements.txt'):
    subprocess.run(['pip', 'install', '-r', 'requirements.txt', '-q'], check=True)
    print('[OK] Dependencies installed.')
```

```python
# Run deterministic evaluation + dashboard rendering
import subprocess, os

# Use the model from the repo (works in Colab after clone)
checkpoint = 'models/best_model_v1.pt'
output = '/content/drive/MyDrive/Snake_AI_PPO/snake_evaluation.mp4'

os.makedirs(os.path.dirname(output), exist_ok=True)

subprocess.run([
    'python', 'evaluate.py',
    '--checkpoint', checkpoint,
    '--output', output,
    '--max_steps', '1000'
], check=True)

print(f'[✓] Video saved to: {output}')
```
